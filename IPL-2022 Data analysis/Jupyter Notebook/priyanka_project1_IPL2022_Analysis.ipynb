# ‚óè	Batting-Most Run Scored by IPL by Teams

```python
# Most Run Scored by IPL by Teams

data=iball.groupby(['BattingTeam'])['total_run'].sum().sort_values(ascending=False)

data.plot(kind='bar',y=data.index,x=data,color='lightblue',edgecolor = 'black',figsize=(15,6))
plt.title(" Most Run Scored by IPL Teams",fontsize=20)
plt.xlabel('BattingTeam',fontsize=18)
plt.xticks(fontsize=15,rotation=80)
plt.ylabel('total_run',fontsize=18)
plt.yticks(fontsize=15)
plt.legend()
plt.show()
```

```python
# Most IPL Runs by a Batsman

plt.figure(figsize=(10,8))

data = iball.groupby(['batter'])['batsman_run'].sum().sort_values(ascending=False)[:10]

sns.barplot(y=data.index,x=data,orient='h',edgecolor = 'black',)
plt.title(" Most Run Scored by IPL Teams (Top 10)",fontsize=20)
plt.xlabel('batsman_run',fontsize=18)
plt.ylabel('batter',fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()

```

# ‚óè	Highest run chase team-wise

```python
hs=iball.groupby(['ID','innings','BattingTeam'])['total_run'].sum().reset_index() 

hs1 = hs[hs['innings']==1]
hs2 = hs[hs['innings']==2]


hs1=hs1.merge(hs2[['ID','innings','total_run','BattingTeam']], on='ID')

hs1.rename(columns={'innings_x':'innings_1','innings_y':\
                    'innings_2','total_run_x':'inning1_runs','total_run_y':'inning2_runs'},inplace=True)


hs1['is_score_chased']=1
hs1['is_score_chased'] = np.where(hs1['inning1_runs']<=hs1['inning2_runs'],'yes', 'no')

hs1[hs1['is_score_chased']=='yes'].sort_values('inning2_runs',axis=0,ascending=False)

```

# ‚óè	Highest Run scorer in power play and death overs

```python
# Highest Runs scored in death overs play by Teams 
fig=iball[iball['overs'].isin ([15,16,17,18,19])].groupby(['BattingTeam']).agg({'total_run':['sum']})


fig.plot(kind='bar',figsize=(12,6),edgecolor = 'black',stacked=True,color='lightpink')
plt.title("Highest Runs scored in death overs play by Teams ",fontsize=20)
plt.xlabel('BattingTeam',fontsize=18)
plt.ylabel('Total Run',fontsize=18)
plt.xticks(fontsize=15,rotation=80)
plt.yticks(fontsize=15)
plt.legend()
plt.show()
```

```python
# Highest Runs scored in power play by Teams 
data=iball[iball['overs'].isin ([0,1,2,3,4,5])].groupby('BattingTeam')['total_run'].sum().sort_values(ascending=False)

plt.figure(figsize=(6,6))
sns.barplot(y=data.index,x=data,orient='h',edgecolor = 'black',palette="Paired")
plt.title('Highest Run scorer in power play',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('BattingTeam',fontsize=18)
plt.ylabel('Total Runs',fontsize=18)
plt.show()
```

# ‚óè	The Lowest total defended and Highest total chased by each team

```python
# The Lowest total defended Match-wise
hs=iball.groupby(['ID','innings','BattingTeam'])['total_run'].sum().reset_index() 

hs1 = hs[hs['innings']==1]
hs2 = hs[hs['innings']==2]


hs1=hs1.merge(hs2[['ID','innings','total_run','BattingTeam']], on='ID')

hs1.rename(columns={'innings_x':'innings_1','innings_y':'innings_2',\
                    'total_run_x':'inning1_runs','total_run_y':'inning2_runs'},inplace=True)

hs1['lowest_total_def']=np.where(hs1['inning1_runs']>hs1['inning2_runs'],'yes', 'no')

hs1[hs1['lowest_total_def']=='yes'].sort_values('inning1_runs',axis=0,ascending=True)
```

```python
# Team wise
fig=hs1[hs1['lowest_total_def']=='yes'].groupby(["BattingTeam_x"])['inning1_runs'].nsmallest(1)

mycolors = ["c", "hotpink", "b", "#4CAF50",'#ff9999','#66b3ff','#99ff99','#ffcc99','Magenta','g']

fig.plot(kind='pie',figsize=(10,8),fontsize=15,\
         explode=[0, 0,0.2,0,0,0,0,0.2,0],startangle=0,colors=mycolors,\
         subplots=True,radius=1,shadow=True,ylabel='',labeldistance=1.08,\
        autopct=lambda x: '{:.0f}'.format(x*fig.sum()/100))

plt.title("The Lowest total defended by each team",fontsize=20)
plt.xlabel('Teams',fontsize=18)
plt.show()

```

```python
# Highest total chased by each team -Match wise
hs=iball.groupby(['ID','innings','BattingTeam'])['total_run'].sum().reset_index() 

hs1 = hs[hs['innings']==1]
hs2 = hs[hs['innings']==2]


hs1=hs1.merge(hs2[['ID','innings','total_run','BattingTeam']], on='ID')

hs1.rename(columns={'innings_x':'innings_1','innings_y':'innings_2','total_run_x':\
                    'inning1_runs','total_run_y':'inning2_runs'},inplace=True)

hs1['is_score_chased']=1
hs1['is_score_chased'] = np.where(hs1['inning1_runs']<=hs1['inning2_runs'],'yes', 'no')

hs1[hs1['is_score_chased']=='yes'].sort_values('inning2_runs',axis=0,ascending=False)
```

```python
# Team wise
fig=hs1[hs1['is_score_chased']=='yes'].groupby(["BattingTeam_y"])['inning2_runs'].nlargest(1)

mycolors = ['#66b3ff',"c",'Magenta','g', '#ff9999','#99ff99',"hotpink", "b", "#4CAF50",'#ffcc99']

fig.plot(kind='pie',figsize=(10,8),fontsize=15,\
         explode=[0, 0,0,0,0.2,0,0,0,0,0],startangle=0,colors=mycolors,\
         subplots=True,radius=1,shadow=True,ylabel='',labeldistance=1.08,\
        autopct=lambda x: '{:.0f}'.format(x*fig.sum()/100))

plt.title("Highest total chased by each team",fontsize=20)
plt.xlabel('Teams',fontsize=18)
plt.show()

```
